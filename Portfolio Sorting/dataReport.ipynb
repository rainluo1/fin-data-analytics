#### 3. Quintile Portfolio

The long short portfolio is constructed as following:

Q1 is the bottom 20% of the quantile, Q5 is the top 20%.
1. ep = Q1 - Q5 
2. lnsize = Q1 -Q5 
3. bk2mkt = Q5 - Q1
4. MOM = Q5 - Q1
5. Beta = Q1 - Q5
6. Ivol = Q5 - Q1

This construction method follows the theory proposed by emperical studies. My data handeling techniques does not contain any look ahead biases, as I am cuting the data into quintile based on its current factors. No future variables are taken into account when performing this decision.
